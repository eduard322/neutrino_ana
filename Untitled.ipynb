```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np
from pathlib import Path
from lhereader import LHEReader


file_list = list(map(str, Path().glob("./DP_electron/interaction_electron/Events/run_*/*.lhe")))

global_list = []
for file in file_list:
    reader = LHEReader(file)
    params = ["eventnumber", "px", "py", "pz", "energy", "mass"]
    print(f"reading {file}...")
    for iev, event in enumerate(reader):
    #     print(event.particles, event.particles.pdgid)
    #     break
        dm_in = list(filter(lambda x: abs(x.pdgid)== 5000521 and x.parent == -1, event.particles))
        dm_out = list(filter(lambda x: abs(x.pdgid)== 5000521 and x.parent == 0, event.particles))
        el_recoil = list(filter(lambda x: abs(x.pdgid)== 11 and x.parent == 0, event.particles))
        dm_in_data = [iev] + [eval(f"dm_in[0].{par}") for par in params[1:]]
        dm_out_data = [eval(f"dm_out[0].{par}") for par in params[1:]]
        el_recoil = [eval(f"el_recoil[0].{par}") for par in params[1:]]
        global_list.append(dm_in_data + dm_out_data + el_recoil)

df = pd.DataFrame(global_list)
df.columns = [params[0]] + [x + "_dm_in" for x in params[1:]] + [x + "_dm_out" for x in params[1:]] + [x + "_el_recoil" for x in params[1:]]
df
```

```python
df = df.sort_values(by="mass_dm_out")
df
```

```python
mass_list = np.unique(df["mass_dm_out"])

fig, ax = plt.subplots(1, 2, figsize = (12,6), dpi = 100)

for mass in mass_list:
    df_mass = df.query(f"mass_dm_out == {mass}")
    ax[0].hist(df_mass["energy_el_recoil"], bins = np.linspace(0, 20, 100), histtype = "step", label = f"DM mass = {mass*1e3:.2f} MeV")
    ax[1].hist(np.arccos(df_mass["pz_el_recoil"]/np.sqrt(df_mass["px_el_recoil"]**2 + df_mass["py_el_recoil"]**2 + df_mass["pz_el_recoil"]**2))*180/np.pi, 
               bins = np.linspace(0, 20, 100), histtype = "step", label = f"DM mass = {mass*1e3:.2f} MeV")
ax[0].set_xlabel("Recoiled electron energy [GeV]")
ax[1].set_xlabel("Recoiled electron angle")
ax[0].set_title("Recoiled electrons. Energy")
ax[0].set_yscale("log")
ax[1].set_title("Recoiled electrons. Angle")
ax[1].set_yscale("log")
ax[1].legend()    
fig.savefig("electrons.pdf")
```
